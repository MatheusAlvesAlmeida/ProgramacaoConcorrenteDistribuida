```python
df = pd.read_csv('consumer256.csv')['1398'].values.tolist()
df2 = pd.read_csv('consumer512.csv')['1159'].values.tolist()
df3 = pd.read_csv('consumer1024.csv')['691'].values.tolist()
df4 = pd.read_csv('consumer2048.csv')['1121'].values.tolist()
df5 = pd.read_csv('consumer5096.csv')['1348'].values.tolist()
```

```python
#average of all values of df
avg = sum(df)/len(df)
avg2 = sum(df2)/len(df2)
avg3 = sum(df3)/len(df3)
avg4 = sum(df4)/len(df4)
avg5 = sum(df5)/len(df5)
```

```python
#standard deviation of all values of df
std = np.std(df)
std2 = np.std(df2)
std3 = np.std(df3)
std4 = np.std(df4)
std5 = np.std(df5)
```

```python
#plot the bar code graph with the average and standard deviation
plt.bar(0, avg, align='center', capsize=10)
plt.bar(1, avg2, align='center', capsize=10)
plt.bar(2, avg3, align='center', capsize=10)
plt.bar(3, avg4, align='center', capsize=10)
plt.bar(4, avg5, align='center', capsize=10)
plt.xticks([0,1,2,3,4], ['256', '512', '1024', '2048', '5096'])
plt.ylabel('Tempo médio de publicação (ms)')
plt.title('Tamanho da mensagem (bytes)')
```
